```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('car_dekho.csv')
print(df.to_string())
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Fuel_Type'].value_counts()
```

```python
fuel=df['Fuel_Type'].value_counts()
colors=['green','orange','red']
fuel.plot(kind='bar',color=colors)
plt.title("Fuel Distribution")
plt.ylabel("car count")

plt.plot()
```

```python
fuel=df['Fuel_Type'].value_counts()
fuel.plot(kind='pie')
plt.title("Fuel Distribution")
plt.plot()
```

```python
df['Transmission'].value_counts()
```

```python
fuel=df['Transmission'].value_counts()
colors=['violet','skyblue']
fuel.plot(kind='bar',color=colors)
plt.title("Transmission Type")
plt.ylabel("car ")
plt.plot()
```

```python
year=df['Year'].value_counts()
colors=['red','blue','green','yellow','orange','purple']
year.plot(kind='bar',color=colors)
plt.title("Years")
plt.ylabel("car ")
plt.plot()
```

```python
year=df['Year'].value_counts()
colors=['red','blue','green','yellow','orange','purple']
year.plot(kind='pie',color=colors)
plt.title("Years")
plt.legend()
plt.ylabel("car ")
plt.plot()
```

```python
seller=df['Seller_Type'].value_counts()
color=['pink','black']
seller.plot(kind='pie',colors=color)
plt.title("Seller Type")
plt.legend()
plt.ylabel("car ")
plt.plot()
```
