```python
#EDA helps us understand patterns, relationships, and outliers in data.
import pandas as pd
import seaborn as sns
df=sns.load_dataset('titanic')
print(df.head())
print(df.info())
```

```python
print(df.describe())  # Summary statistics
print(df['age'].mean())  # Mean age of passengers
print(df['survived'].value_counts())  # Count of survived vs. not survived
```

```python
import matplotlib.pyplot as plt

sns.boxplot(x=df["age"])
plt.title("Age Distribution")
plt.show()
```

```python
print(df.dtypes)  # Check data types of each column

```

```python
df['sex'] = df['sex'].map({'male': 0, 'female': 1})  # Male = 0, Female = 1
df['embarked'] = df['embarked'].map({'C': 0, 'Q': 1, 'S': 2})  # Convert 'C', 'Q', 'S' to numbers
```

```python
df['sex'] = df['sex'].map({'male': 0, 'female': 1})  # Male → 0, Female → 1
df['embarked'] = df['embarked'].map({'C': 0, 'Q': 1, 'S': 2})  # C → 0, Q → 1, S → 2
df['alive'] = df['alive'].map({'no': 0, 'yes': 1})  # No → 0, Yes → 1
df['alone'] = df['alone'].astype(int)  # Convert True/False to 1/0
df['adult_male'] = df['adult_male'].astype(int)  # Convert True/False to 1/0
```

```python
df_numeric = df.select_dtypes(include=['int64', 'float64'])  # Keep only numeric columns
```

```python
corr = df_numeric.corr()  # Compute correlation on numeric data
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
