```python
x=np.array([6,8,8,9])
y=np.array([1,10,2,9])
z=plt.plot(x,y,linestyle='dotted',linewidth=2,
          color='red',marker="*",markersize=20,markerfacecolor="yellow",markeredgecolor="blue")
print(z)
```

```python
x1=np.array([2,5,6])
y1=np.array([3,4,6])
x2=np.array([3,5,4])
y2=np.array([2,7,4])
plt.plot(x1,y1,marker="o",markersize=8)
plt.plot(x2,y2,marker="o",markersize=8)
plt.grid()
plt.show()

```

```python
x1=np.array([2,5,6])
y1=np.array([3,4,6])
x2=np.array([3,5,4])
y2=np.array([2,7,4])
plt.plot(x1,y1,marker="o",markersize=8)
plt.plot(x2,y2,marker="o",markersize=8)
plt.grid(axis='x',linestyle="dotted",color='green',linewidth=1
        )
plt.title("Analysing Data",loc='left')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#student data and marks
marks=np.array([40,50,80,70,60,30,50,10,95,67,43])
student=np.array([5,4,8,19,1,2,3,7,6,13,1])
colors=["red","yellow"]
plt.barh(marks,student,color=colors)

plt.show()
```

```python
m=np.array([40,50,80,70,60])
s=np.array([5,4,8,19,1])
m1=np.array([4,5,6,7,9])
s1=np.array([6,5,6,1,9])
colors=["red","yellow","green","pink","black"]
plt.scatter(m,s,color=colors)
plt.scatter(m1,s1)
plt.show()
```

```python
#pie chart
a=np.array([5,3,2,9,8])
colors=["red","black","purple","yellow","pink"]
label=["red","black","purple","yellow","pink"]
plt.pie(a,colors=colors,labels=label)
plt.legend()
plt.show()
```

```python
#Histogram
x2=np.array([3,5,4,8,9,7])
plt.hist(x2)
plt.grid()
plt.show()
```

```python
#use random 
from numpy import random
a=np.array([1,2,3,99,8,7])
b=random.randint(a)
plt.hist(b)
plt.show()
```
